### Instalaciones y factores útiles

#### Utilizar GPU propia
---

Para ésto deberá instalar __[Cuda](https://developer.nvidia.com/cuda-toolkit-archive)__ en su computador. Cuda es una librería en $C$ para utilizar la GPU de manera paralela. La versión a instalar dependerá fuertemente de las librerías que utilizaremos y su versión. Cuda 9 es la que tiene mayor soporte y compatibilidad con librerías, sin embargo puede consultar las dependencias de las versiones de librerías, por ejemplo la de tensorflow-gpu: https://www.tensorflow.org/install/source#tested_source_configurations


Para verificar que *Cuda* está instalado en su computador y tiene las dependencias, deberá ejecutar:
```
nvcc --version
```

Para verificar la información, si es que su computador tiene, de la *GPU* deberá ejecutar:
```
nvidia-smi
```

#### Para instalar librerías
---

Acceder al ambiente creado
```
source activate myenv
```
ó
```
conda activate myenv
```

* Vía pip (+*Colaboratory*)
    ```
    python -m pip install jupyter numpy scipy seaborn ipykernel pandas matplotlib sklearn
    ```
    > Librerías avanzadas de data science:
    ```
    python -m pip install tensorflow-gpu==version keras
    ```
    
    Afortunadamente *Colaboratory* viene con muchas librerías compatibles ya instaladas, pero si desea instalar o cambiar de versión alguna puede ejecutar:    
    ```
    !pip install libreria==version
    ```   

* Vía conda
    ```
    conda install jupyter numpy scipy seaborn ipykernel pandas matplotlib sklearn 
    ```
    > Librerías avanzadas de data science:
    ```
    conda install tensorflow-gpu==version keras-gpu
    ```
    
* Abrir un Jupyter
    ```
    jupyter notebook
    ```
    
Para salir del entorno
```
source/conda deactivate
```

*No olvide revisar la compatibilidad de las librerías entre sí o  con Cuda (si es que va a utilizar GPU).*

## Buenas prácticas y tips
---

Algunos *tips* útiles del framework Jupyter.

* **ctrl + shift + p** = abre buscador de comandos  
* **ctrl + shift + -** = divide la celda actual desde donde se encuentra el cursor  
* **Esc + f** = find and replace en código sin afectar salidas  
* **shift + M** = merge de celdas
